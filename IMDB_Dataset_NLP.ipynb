```python
def chat_conversation(text3):
  new_text = []
  for w in text3.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

##Handling emojis


```python
import re
def remove_emoji(text5):
  emoji_pattern = re.compile("["
                          u"\U0001F600-\U0001F64F"  # emoticons
                          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                          u"\U0001F680-\U0001F6FF"  # transport & map symbols
                          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                          u"\U00002702-\U000027B0"
                          u"\U000024C2-\U0001F251"
                          "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', text5)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations = "?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
  if word in punctuations:
    sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
  print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```

```python
story
```

- Window Size (window=10): Determines how many words before and after a target word are considered for learning relationships.
- Minimum Count (min_count=2): Ignores words that appear less than twice in the dataset.

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,

)
```

```python
#to extract the unique words
model.build_vocab(story)
```

- Now, the model learns word relationships by training on story. It tries to adjust word vectors so that words appearing in similar contexts have similar embeddings.


**How Word2Vec Learns Word Relationships ?**
Word2Vec uses neural networks to learn word relationships based on context. It uses one of two approaches:

- CBOW (Continuous Bag of Words) → Predicts a word given its surrounding words.
- Skip-gram → Predicts surrounding words given a target word.
- Your model is using one of these techniques (by default, it's Skip-gram if not specified).



```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

## To increase the accuracy -> Use RandomForestClassifier


-What is RandomForestClassifier()?
**Random Forest is an ensemble learning method that combines multiple Decision Trees to improve prediction accuracy and reduce overfitting.**

- Unlike Naïve Bayes (which assumes feature independence), Random Forest:
- Creates multiple Decision Trees on different subsets of data.
- Combines their outputs (majority vote for classification).
- Handles non-linearity and interactions between words better than Naïve Bayes.


```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_bow,Y_train)     #Train the model on training data

y_pred = rf.predict(X_test_bow)  # Predict the lables for the test data
accuracy_score(Y_test,y_pred)
```

- What is a Decision Tree?
**A Decision Tree is a supervised learning algorithm used for classification and regression tasks. It works like a flowchart, making decisions based on data features step by step.**

Think of it as a series of "Yes" or "No" questions that lead to a final decision.

```python
                #  (Root Node)
                #      "worst" in review?
                #     /        \
                # Yes (0)     No
                #          "amazing" in review?
                #         /        \
                #    Yes (1)     No
                #             "not" in review?
                #             /       \
                #         Yes (0)    No (1)


cv = CountVectorizer(max_features = 3000)
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,Y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(Y_test,y_pred)
```

##Average word2vec

- Calculate the average Word2Vec vector for a document (review).

```python
def document_vector(doc):
  #remove out-of-vocabulary words
  doc = [word for word in doc.split() if word in model.wv.index_to_key]
  return np.mean(model.wv[doc],axis=0)
```
